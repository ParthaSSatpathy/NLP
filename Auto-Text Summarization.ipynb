```python
# Putting the whole code at one Place
import urllib.request as ur # To scrap text from webpage
from bs4 import BeautifulSoup # To parse the webpage
import re
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem.lancaster import LancasterStemmer
from nltk.probability import FreqDist
from collections import defaultdict
from heapq import nlargest
from collections import defaultdict

def retText(URL):
    page = ur.urlopen(URL).read().decode('utf8','ignore')
    soup = BeautifulSoup(page,'lxml')
    text = " ".join(map(lambda p:p.text,soup.find_all('article')))
    return text

def preprocess(text):
    text = re.sub(r'\d+', '', text) # Removes all Number from the text
    text = text.replace('“', "") # Removes “ from the text which cannot be removed by nltk
    words = word_tokenize(text.lower())
    customStopWords = set(stopwords.words('english')+list(punctuation))
    words = [word for word in words if word not in customStopWords]
    words = [LancasterStemmer().stem(word) for word in words]
    return words

def createAbstract(text,abstractSize):
    
    sents = sent_tokenize(text)
    words = preprocess(text)
    
    # Let's find the highest frequency words
    freq = FreqDist(words)
    
    # Let's find the ranking of the individual sentences of the text
    ranking = defaultdict(int) # An empty dictionary

    for i,sent in enumerate(sents): # For every sentnce in the text
        for w in preprocess(sent):  # For every word in a specific sentence
            if w in freq:           # If a word exists in teh frequency dictionary
                ranking[i] += freq[w]  # Find the frequency of the word and add it to the rank of the sentence
                
    sents_idx = nlargest(abstractSize,ranking,key=ranking.get) # Top ranked sentences index
    abstract = [sents[idx] for idx in sorted(sents_idx)]
    
    return abstract

def main():
    articleURL = "https://www.washingtonpost.com/politics/federal-court-blocks-trump-epa-on-air-pollution/2017/07/03/464a7344-601e-11e7-84a1-a26b75ad39fe_story.html?hpid=hp_hp-top-table-main_methane-521pm%3Ahomepage%2Fstory&utm_term=.b1e5d2defded"
    myText = retText(articleURL)
    myAbstract = createAbstract(myText,4)
    print(myAbstract)
    
main()
```
